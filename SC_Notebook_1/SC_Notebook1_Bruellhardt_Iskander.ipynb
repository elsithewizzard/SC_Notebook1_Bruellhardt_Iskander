4. Now we want to extract each half day's forecast from the extended forecast on the weather report page. <br>
    a. Find the container for the extended forecast on the weather page we just downloaded. <br>
    b. Make a list with all forecast items (overnight, Wednesday, Wednesday night, etc.) <br>
    c. For each time period, print out the name of the period, the short description of the expected weather conditions, and the temperature. Try to output a nicely formated string.

**Example output for one time period:** <br>
_Overnight, the weather will be mostly clear and breezy and the temperature will be 65F._

````python
import requests
from bs4 import BeautifulSoup

soup = BeautifulSoup(requests.get("https://forecast.weather.gov/MapClick.php?lon=-157.835&lat=21.286").text, "html.parser")

ul = soup.find("ul", class_="list-unstyled")
for li in ul.find_all("li"):
    period = li.find("p", class_="period-name").get_text(separator=" ",strip=True)
    temp = " ".join(li.find("p", class_="temp").get_text(strip=True).split(":")[1].split())
    desc = li.find("p", class_="short-desc").get_text(separator=" ", strip=True)

    print(f"{period}, the weather will be {desc} and the temperature will be {temp}")

````

**Hint**: Use the `headers` argument in your requests for this exercise.

#comment: Teacher is often employed by a canton (e.g. Nidwalden), this is not the location but the employer.

````python
import json
import os

# directory for HTML and json files (to keep our work space clean)
output_dir = "HTML and JSON Files 2b"
os.makedirs(output_dir, exist_ok=True)

five_jobs = ['teacher', 'lawyer', 'scientist', 'banker', 'psychologist']
job_dict = {}

for job in five_jobs:
    #initiate dictionary entry with for Job as key
    job_dict[job]= []
    #1. get the html files
    jobscout_page = requests.get(
    url= f"https://www.jobscout24.ch/de/jobs/{job}", 
    headers= {'User-Agent': "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36"}
    )
    with open(f"{output_dir}/{job}.html", "w", encoding="utf-8") as f:
        f.write(jobscout_page.text)
    soup = BeautifulSoup(jobscout_page.content, 'html.parser')
    #2. Get company names
    companies = soup.find_all('span', class_='company')
    for company in companies:
        name = company.get_text(strip=True)
        #append companies to our job dictionary
        job_dict[job].append(name)

    time.sleep(2)

#3. save json
json_str = json.dumps(job_dict, indent=4)
with open(f"{output_dir}/jobs.json", "w") as f:
    f.write(json_str)

#4. print dictionary
print(job_dict)
````
