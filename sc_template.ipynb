2. Create a Virtual Environment
What’s that?
Think of a Virtual Environment as a dedicated project fridge. Without it, all your Python projects share
one giant, messy communal fridge. Eventually, one project’s old leftovers leak onto your fresh ingredients,
or someone replaces your milk with heavy cream, and your whole recipe is ruined.
Virtual environments make your life easier by giving this project its own clean fridge – or, in
this case, its own clean library space. With an environment, you prevent library conflicts between different
projects and if you make a mess here, it will not spoil your other projects! Another advantage is that once
you have put something in your own fridge, you know that it will still be there the next time you open
the fridge. That means you will not have to install the same libraries again if you need them in future
notebooks, as long as you are always working within the same environment (yes, you might want to update
some ingredients if they are outdated from time to time – which would also be a good concept in some
communal fridges. . . ).
1. Navigate to your project folder: cd path/to/your/folder
2. Create the environment:
• Mac/Linux: python3 -m venv .venv
• Windows: python -m venv .venv
3. Activate the environment (open the fridge):
• Mac/Linux: source .venv/bin/activate
• Windows: .venv\Scripts\activate
• Note: You should now see (.venv) at the start of your terminal line.

3. Stock the Fridge (Install the Required Libraries)
Instead of buying ingredients one by one, we use a requirements.txt “grocery list”. We can then tell our
package manager (pip) to installing all libraries that are listed in the requirements.txt file.
With your environment activated, ensure the requirements.txt file is in your current folder and run:
# Install or update the package installer
pip install --upgrade pip
# Install required libraries
pip install -r requirements.txt
