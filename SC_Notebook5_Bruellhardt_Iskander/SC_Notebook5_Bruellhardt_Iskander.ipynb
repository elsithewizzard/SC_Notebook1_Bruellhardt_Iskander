````python
### Any code you use to evaluate the data automatically here:
def compare_raters(path_file_a, path_file_b, col_name):
    '''Compares the columns for differences and returns a percentage on how similar they are. 
    Only works if the two files have the same col_name and length!
    
    Returns a list with the percentage of overlap, tuple of absolute values(same, different) and a list of indeces where differences occured'''
    same = 0
    different = 0
    different_index = []
    file_a = pd.read_csv(path_file_a)
    col_a = list(file_a[col_name])

    file_b = pd.read_csv(path_file_b)
    col_b = list(file_b[col_name])

    for i in range(len(file_a)):
        if col_a[i].lower() == col_b[i].lower():
            same += 1
        else:
            different += 1
            different_index.append(i)

        #info about interrater differences 
    percentage = f'{int(100/len(file_a)*same)}%'
    list_with_info = [percentage, (same, different), different_index]
    return list_with_info

def compare_cols(path_file, col_a, col_b):
    '''Compares the columns for differences and returns a percentage on how similar they are. 
    Only works if the two files have the same col_name and length!
    
    Returns a list with the percentage of overlap, tuple of absolute values(same, different) and a list of indeces where differences occured'''
    same = 0
    different = 0
    different_index = []
    file_a = pd.read_csv(path_file)
    col_a = list(file_a[col_a])
    col_b = list(file_a[col_b])

    for i in range(len(file_a)):
        if col_a[i].lower() == col_b[i].lower():
            same += 1
        else:
            different += 1
            different_index.append(i)

        #info about interrater differences 
    percentage = f'{int(100/len(file_a)*same)}%'
    list_with_info = [percentage, (same, different), different_index]
    return list_with_info



def assess_coverage(dataframe, column_name):
    '''Compares the columns for differences and returns a percentage on how similar they are. 
    Only works if the two files have the same col_name and length!
    
    Returns a list with the percentage of overlap, tuple of absolute values(same, different) and a list of indeces where differences occured'''
    unclear = 0
    clear = 0
    l = list(dataframe[column_name])
    
    for item in l:
        if item.lower() == 'unclear':
            unclear += 1
        else:
            clear += 1

    #info about coverage 
    total = unclear + clear
    coverage = f'{int(100/total*clear)}%'
    return coverage



cwd = os.getcwd()
elias_path =  os.path.join(os.getcwd(), 'emails_manually_elias.csv')
alina_path = os.path.join(os.getcwd(), 'emails_manually_alina.csv')

#compared to alina
print(f''' 
Coverage = {assess_coverage(data, 'country')}

Accuracy Automatic Detection (compared to Alina) = {compare_cols(alina_path, 'country', 'manual_classifier')[0]} 
Accuracy Automatic Detection (compared to Elias) = {compare_cols(elias_path, 'country', 'manual_classifier')[0]} 

Interrater Overlap  = {compare_raters(alina_path, elias_path, 'manual_classifier')[0]}
''')
````

### Exercise 2.1: Data Preparation (6 points)

Understand the directory structure (check out README.md file present in the downloaded folder) and write pandas code to create a dataframe with the following columns to store all the reviews:

1. **review:** The original review
2. **split_name:** Should have the values `train`, `test`, or `unsup` - determined by the folder the review is in
3. **rating:** Rating of the review
4. **label:** If the review is from `pos`/`neg` folder, `label` should be `pos`/`neg` respectively. If the label is from `unsup` folder, set the value of `label` as `na`

Every row represents a review and every column represents a property of the review. Display the dataframe to verify whether your solution looks correct.

````python
import pandas as pd
import os

base_path = os.path.join(os.getcwd(), 'aclImdb')

def prepare_imdb_data(base_dir):
    data = []
    splits = ['train', 'test']
    
    for split in splits:
        split_path = os.path.join(base_dir, split)
        
        if not os.path.exists(split_path):
            continue

        for folder in os.listdir(split_path):
            folder_path = os.path.join(split_path, folder)
            
            if not os.path.isdir(folder_path):
                continue
                
            label = folder if folder in ['pos', 'neg'] else 'na'
            
            for filename in os.listdir(folder_path):
                if filename.endswith(".txt"):
                    file_full_path = os.path.join(folder_path, filename)
                    
                    try:
                        rating = filename.split('_')[1].split('.')[0]
                    except (IndexError, ValueError):
                        rating = 0

                    with open(file_full_path, 'r', encoding='utf-8') as f:
                        review_text = f.read()
                        
                    data.append({
                        'review': review_text,
                        'split_name': split,
                        'rating': int(rating),
                        'label': label
                    })
                    
    return pd.DataFrame(data)

df = prepare_imdb_data(base_path)
print(df.head())
````

Scale Choice: 1 to 5

To be consistent with the manual rating of Elias, I chose to apply the same rating scale from 1-5 for this LLM prompt. It allows for a more granular sentiment analysis than a binary (0/1) scale, making better use of the LLM capabilities.

Prompt:
Analyze the sentiment of the following movie review. 
Assign a sentiment score using a 5-point scale: 1 = Very Negative, 2 = Negative, 3 = Neutral, 4 = Positive, 5 = Very Positive.
Output ONLY the integer score. Do not provide explanations or extra text.

````python
%pip install -U google-genai
````

````python
from google import genai
import pandas as pd
from dotenv import load_dotenv
import os
import time

load_dotenv()
client = genai.Client(api_key=os.getenv('GEMINI_KEY'))

df_imdb_random_reviews = pd.read_csv('imdb_random_reviews.csv')

prompt_template = (
    """Analyze the sentiment of the following movie review. 
    Assign a sentiment score using a 5-point scale: 1 = Very Negative, 2 = Negative, 3 = Neutral, 4 = Positive, 5 = Very Positive.
    Output ONLY the integer score. Do not provide explanations or extra text. 
    Review: {review_text}"""
)

annotations = []

for index, row in df_imdb_random_reviews.iterrows():
    if index >= 30: 
        break
    
    try:
        response = client.models.generate_content(
            model="gemini-2.5-flash", #is that the best free-tier one? 2.0 didn't work!
            contents=prompt_template.format(review_text=row['review'])
        )
        annotations.append(response.text.strip())
        time.sleep(8) # to stay savely under 15 requests / minute limit
        
    except Exception as e:
        if "429" in str(e): #rate limit exceeded
            print(f"Error at index {index}: {e}")
            break 
        else:
            print(f"Error at index {index}: {e}")
            annotations.append('N/A')

df_llm = df_imdb_random_reviews.copy()
remaining = 30 - len(annotations) 
annotations.extend(['N/A'] * remaining)

df_llm['label'] = annotations
df_llm.to_csv('imdb_llm_alina.csv', index=False)
````

**Explain which value you use for the annotation, what its meaning is and justify your choice:**

The compound score is a single value ranging from -1 (most negative) to +1 (most positive), aggregating VADER's positive, negative, and neutral sub-scores. It's the standard output for overall sentiment and gives a continuous scale that can be binned into the 1-5 categories later.

````python
### Your code here:
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
random_reviews = pd.read_csv('imdb_random_reviews.csv')
random_reviews['annotation'] = random_reviews['review'].apply(lambda r: sia.polarity_scores(r)['compound'])
random_reviews.to_csv('imdb_nltk.csv', index=False)
random_reviews[['label', 'annotation']]
````

### Exercise 2.4: Normalizing the Different Scales (5 points)

Since we used different scales in different types of annotations, we need to convert all of them to the same scale for comparison.

**Your tasks:**
- Convert the NLTK scores to the Likert-style scale used for human annotation.
- Do the same for the LLM-based scores, if you did not already use the same scale as for human annotation.
- Mention how you choose any thresholds and any potential pitfalls of the methodology.
- Create a new dataframe based on the random sample dataframe from the preparation task of exercise 2.2, adding three columns for the normalized annotations to it. This dataframe should contain the following columns:
    1. **review**
    2. **split_name**
    3. **rating**
    4. **label**
    5. **human:** Scores from human annotation
    6. **llm:** Normalized scores from LLM-based annotation
    7. **nltk:** Normalized scores from NLTK-based annotation

**Output:**

Store the dataframe in a CSV file named `imdb_combined.csv` and include it in your submission.

````python
### Your code for scale conversions here:
#loading all datasets
df_human = pd.read_csv('imdb_human.csv')
df_llm = pd.read_csv('imdb_llm.csv')
df_nltk = pd.read_csv('imdb_nltk.csv')
df_normalized = pd.read_csv('imdb_random_reviews.csv').copy()


human_annotations = df_human['annotation'] 
llm_annotations = df_llm['annotation'] 
nltk_annotations = df_nltk['annotation'] 

normalized_human = [int(x) for x in human_annotations]
normalized_llm = [int(x) if not pd.isna(x) else 'NaN' for x in llm_annotations]
normalized_nltk = [int(round(x*2 + 3, 0)) for x in nltk_annotations]



df_normalized['human'] = normalized_human
df_normalized['llm'] = normalized_llm
df_normalized['nltk'] = normalized_nltk

df_normalized.to_csv('imdb_combined.csv')
df_normalized.head(30)
````

**You answer here:**
So our human and LLM annotations already matched the scale from 1-5, to convert the -1 to 1 scale from the nltk, we used the following logic:

y = 2x + 3 

This is because the range from -1 to 1 = 3, the range of 1 to 5 = 4. We therefore multiplied the nltk values by 2.
To shift the values to start at 1 and be neutral at 3 we added 3 (0 * 2 + 3 = 3).

⮕ We then rounded(y) to be the closest integer using round() which returns the nearest integer. We realize that rounding loses a little bit of the granularity from the float aspect, however, integers are much easier to interpret and the task required a likert-scale.
